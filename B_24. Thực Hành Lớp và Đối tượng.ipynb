Bài 1:

```python
""" Viết Class NhanVien gồm các thuộc tính :
+ Tên 
+ Tuổi 
+ Địa chỉ 
+ Tổng số giờ làm

Và các phương thức :
- Phương thức tạo
- input_info(): Nhập các thông tin cho nhân viên từ bàn phím
- print_info(): In ra tất cả các thông tin của nhân viên
- tinh_thuong(): Tính toán và trả về số tiền thưởng của nhân viên theo công thức sau :
Nếu tổng số giờ làm của nhân viên >= 200 thì thưởng = lương * 20%
Nếu tổng số giờ làm của nhân viên <200  và >= 100 thì thưởng = lương * 10%
Nếu tổng số giờ làm của nhân viên <100 thì thưởng = 0"""
```

```python
class NhanVien():
    def __init__(self, name, age, address, work_hour):
        self.name = name
        self.age = age
        self.address = address
        self.work_hour = work_hour
    
    def input_info(self):
        self.name = input('Nhập tên nhân viên: ')
        self.age = input('Nhập tuổi nhân viên: ')
        self.address = input('Nhập địa chỉ nhân viên: ')
        self.work_hour = int(input('Nhập tổng số giờ làm của nhân viên: '))
        
    def print_info(self):
        print('Tên:',self.name)
        print('Tuổi: ',self.age)
        print('Địa chỉ:',self.address)
        print('Tổng số giờ làm:',self.work_hour)
        
    def tinh_thuong(self, salary):
        if self.work_hour >= 200:
            return salary*1.2
        elif 100 <= self.work_hour < 200:
            return salary*1.1
        return 0
```

```python
"""Tạo Class có thên SoHoc gồm có các thuộc tính và phương thức sau :
+ Thuộc tính : number1 ,number2
+ Phương thức :
- phương thức tạo __int()__
- input_info(): dùng để nhập 2 số number1 và number2
- print_info(): dùng để hiển number1 và number2
- addition(): dùng để cộng number1,number2
- subtract(): dùng để trừ number1 và number2
- multi(): dùng để nhân number1 và number2 
- division(): dùng để chia number1 và number2

Sau khi tạo class , khởi tạo 1 object SoHoc và in ra kết quả tính toán từ các hàm thực hiên phép tính"""
```

```python
class SoHoc:
    def __init__(self, number1, number2):
        self.number1 = number1
        self.number2 = number2
    
    def input_info(self):
        self.number1 = int(input('Nhập số thứ nhất: '))
        self.number2 = int(input('Nhập số thứ hai: '))

    def print_info(self):
        print(f'Number 1: {self.number1}')
        print(f'Number 2: {self.number2}')

    def addition(self):
        return self.number1 + self.number2
    
    def subtract(self):
        return self.number1 - self.number2
    
    def multi(self):
        return self.number1 * self.number2
    
    def division(self):
        return self.number1 / self.number2
    
obj = SoHoc(0,0)
obj.input_info()
print(obj.addition())
print(obj.subtract())
print(obj.multi())
print(obj.division())
```

```python
"""Hãy tạo class SinhVien gồm các thông tin: họ tên, mã số, ngày sinh, địa chỉ, điểm
trung bình 4 năm học, số điện thoại. Mỗi sinh viên có thể có tối đa 3 số điện thoại.
Viết chương trình cho phép:
- Nhập một danh sách các sinh viên với đầy đủ các thông tin nêu trên.
- Thêm một sinh viên vào danh sách.
- Tìm kiếm sinh viên trong danh sách theo mã số hoặc theo họ tên.
- Xóa một sinh viên khỏi danh sách dựa theo mã số.
- Sửa dữ liệu một sinh viên.
- Sắp xếp lại các sinh viên theo họ tên hoặc theo mã số.
- Sắp xếp lại các sinh viên theo điểm trung bình từ cao đến thấp.
- Xuất danh theo thứ tự mã số tăng dần.
- Xuất danh sách theo thứ tự điểm trung bình giảm dần và có thêm cột sắp hạng
sinh viên. 
 """
```

```python
class SinhVien:
    
    def __init__(self,name,mssv,dob,address,gpa,sdt1,sdt2='không có',sdt3='không có'):
        self.name=name
        self.mssv=mssv
        self.dob=dob
        self.address=address
        self.gpa=gpa
        self.sdt1=sdt1
        self.sdt2=sdt2
        self.sdt3=sdt3
    
    def show(self):
        print('Tên: %s, MSSV: %s, DoB: %s, Địa chỉ: %s, GPA: %s, Sđt 1: %s, Sđt 2: %s, Sđt 3: %s' \
              %(self.name,self.mssv,self.dob,self.address,self.gpa,self.sdt1,self.sdt2,self.sdt3))    

def add_SV(lst):
    name=input('Nhập tên SV: ')
    mssv=input('Nhập MSSV: ')
    dob=input('Nhập ngày tháng năm sinh: ')
    address=input('Nhập địa chỉ SV: ')
    gpa=input('Nhập điểm trung bình 4 năm: ')
    sdt1=input('Nhập số điện thoại 1 của SV: ')
    sdt2=input('Nhập số điện thoại 2 của SV, nhập \'n\' nếu không có: ')
    if sdt2=='n':
        sdt2='không có'
        sdt3='không có'
    else:
        sdt3=input('Nhập số điện thoại 3 của SV, nhập \'n\' nếu không có: ')
        if sdt3=='n':
            sdt3='không có'
    sv = SinhVien(name,mssv,dob,address,gpa,sdt1,sdt2,sdt3)
    lst.append(sv)

def find_SV(lst,name):
    check=False
    for i in range(len(lst)):
        if lst[i].name==name or lst[i].mssv==name:
            lst[i].show()
            check=True
    if check==False:
        print('SV không có trong danh sách')

def del_SV(lst,mssv):
    check=False   
    for i in range(len(lst)):
        if lst[i].mssv==mssv:
            check=True
            print('Bạn đang xóa SV %s' %(lst[i].name))
            while True:
                choice=input('Bạn có muốn tiếp tục không? (y/n): ')
                if choice.lower() in ['y','n']:
                    break
                else:
                    print('Bạn nhập sai cú pháp!')
            if choice.lower()=='y':
                del lst[i]
                print('Xóa thành công!')
    if check==False:
        print('SV không có trong danh sách')

def adjust_SV(lst,mssv):
    check=False
    for i in range(len(lst)):
        if lst[i].mssv==mssv:
            print('Bạn đang chỉnh sửa thông tin SV: ',lst[i].name)
            sv=lst[i]
            check=True
            break
    if check==False:
        print('SV không có trong danh sách')
    else:
        choice=''
        while True:
            choice=input('Bạn có muốn tiếp tục không? (y/n): ')
            if choice.lower() in ['y','n']:
                break
            else:
                print('Bạn nhập sai cú pháp!')
        if choice=='y':
            sv.name=input('Nhập tên SV: ')
            sv.mssv=input('Nhập MSSV: ')
            sv.dob=input('Nhập ngày tháng năm sinh: ')
            sv.address=input('Nhập địa chỉ SV: ')
            sv.gpa=input('Nhập điểm trung bình 4 năm: ')
            sv.sdt1=input('Nhập số điện thoại 1 của SV: ')
            sv.sdt2=input('Nhập số điện thoại 2 của SV, nhập \'n\' nếu không có: ')
            if sv.sdt2=='n':
                sv.sdt2='không có'
                sv.sdt3='không có'
            else:
                sv.sdt3=input('Nhập số điện thoại 3 của SV, nhập \'n\' nếu không có: ')
                if sv.sdt3=='n':
                    sv.sdt3='không có'
                    
def sx_lst_mssv(lst):
    sorted(lst,key=lambda x:x.mssv)
    
def sx_lst_gpa(lst,reverse=True):
    sorted(lst,key=lambda x:x.gpa,reverse=reverse)

def show_lst(lst):
    for i in range(len(lst)):
        lst[i].show()

def con():
    global run
    while True:
        choice=input('Bạn có muốn thực hiện tiếp không? (y/n): ')
        if choice.lower()=='n':
            run=False
            break
        elif choice.lower()=='y':
            break
        else:
            print('Bạn nhập sai cú pháp!')
            
lst_SV=[]
run=True
while run:
    print('Option: \n\
    1. Thêm SV vào danh sách\n\
    2. Tìm kiếm SV theo MSSV hoặc họ và tên\n\
    3. Xóa SV khỏi danh sách theo MSSV\n\
    4. Sửa dữ liệu SV theo MSSV\n\
    5. Xuất danh sách theo MSSV tăng dần\n\
    6. Xuất danh sách theo GPA tăng dần\n\
    7. Xuất danh sách theo GPA giảm dần\n\
    8. Thoát')
    choice=input('Lựa chọn của bạn: ')
    if choice=='1':
        add_SV(lst_SV)
        con()
    elif choice=='2':
        name=input('Nhập MSSV hoặc họ tên của SV bạn muốn tìm: ')
        find_SV(lst_SV, name)
        con()
    elif choice=='3':
        mssv=input('Nhập MSSV của SV bạn muốn xóa: ')
        del_SV(lst_SV, mssv)
        con()
    elif choice=='4':
        mssv=input('Nhập MSSV của SV bạn muốn chỉnh sửa: ')
        adjust_SV(lst_SV, mssv)
        con()
    elif choice=='5':
        sx_lst_mssv(lst_SV)
        show_lst(lst_SV)
        con()
    elif choice=='6':
        sx_lst_mssv(lst_SV)
        show_lst(lst_SV)
        con()
    elif choice=='7':
        sx_lst_gpa(lst_SV,False)
        show_lst(lst_SV)
        con()
    elif choice=='8':
        run=False
    else:
        print('Bạn nhập sai cú pháp!')
```

```python
"""
Xây dựng lớp Phân số (Fraction) có các thành phần sau:
- Các thuộc tính: Tử số, mẫu số.
- Phương thức tạo để khởi tạo giá trị cho tử số và mẫu số.
- Các phương thức setter và getter.
- Các phương thức:
    + Nhập phân số
    + In Phân số
    + Rút gọn phân số
    + Nghịch đảo phân số
    + add(), sub(), mul(), div() tương ứng để thực hiện cộng, trừ, nhân, chia hai phân số cho nhau.
"""
```

```python
class PhanSo():
    
    def __init__(self, a, b):
        self.__a = a
        self.__b = b
        if b == 0:
            raise TypeError('Denominator cann\'t be zero')
    
    @property
    def a(self):
        return self.__a

    @a.setter
    def a(self, a):
        self.__a = a

    @property
    def b(self):
        return self.__b

    @b.setter
    def b(self, b):
        self.__b = b

    def nhap_ps(self):
        self.a = int(input('Nhập tử số: '))
        self.b = int(input('Nhập mẫu số: '))
    
    def in_ps(self):
        print(self)

    def __str__(self):
        return f'{self.a}/{self.b}' 
    
    def rut_gon(self):        
        a = self.a
        b = self.b
        ucln = 1
        for i in range(1,min(abs(a),abs(b))+1):
            if i > ucln and a%i==0 and b%i==0:
                ucln=i
        a = a//ucln
        b = b//ucln
        return PhanSo(a,b)
    
    def nghich_dao(self):
        return PhanSo(self.b, self.a)
    
    def __add__(self,o):
        a = self.a*o.b + self.b*o.a
        b = self.b*o.b
        return PhanSo(a,b).rut_gon()
    
    def __sub__(self,o):
        a = self.a*o.b - self.b*o.a
        b = self.b*o.b
        return PhanSo(a,b).rut_gon()
    
    def __mul__(self,o):
        a = self.a * o.a
        b = self.b * o.b
        return PhanSo(a,b).rut_gon()
    
    def __truediv__(self,o):
        a = self.a * o.b
        b = self.b * o.a
        return PhanSo(a,b).rut_gon()
```

```python
"""Xây dựng lớp hình chữ nhật (Rectangle) có các thành phần sau:
* Các thuộc tính: chiều dài, chiều rộng.
* Các phương thức:
- Nhập chiều dài, chiều rộng
- Tính diện tích
- Tính chu vi"""
```

```python
class Rectangle():

    def __init__(self, chieu_dai, chieu_rong):
        self.chieu_dai = chieu_dai
        self.chieu_rong = chieu_rong

    def input_rec(self):
        self.chieu_dai = int(input('Nhập chiều dài: '))
        self.chieu_rong = int(input('Nhập chiều rộng: '))

    def dien_tich(self):
        return self.chieu_rong * self.chieu_dai
    
    def chu_vi(self):
        return (self.chieu_dai + self.chieu_rong) * 2
```
